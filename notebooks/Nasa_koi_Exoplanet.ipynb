```python
# Total_score : 70
koi_yes['koi_cat_score'] = ""

# Iterate over 'koi_steff' values and assign scores to total 100
for index, i in koi_yes['koi_steff'].items():  # Iterate using index and value
    if i > 1000:
        koi_yes.loc[index, 'koi_cat_score'] = 30  # Score for "Hot Planet"
    elif (300 <= i < 1000):
        koi_yes.loc[index, 'koi_cat_score'] = 50  # Score for "Warm Planet"
    elif (200 <= i < 300):
        koi_yes.loc[index, 'koi_cat_score'] = 70  # Score for "Habitable Zone Exoplanet"
    elif i < 200:
        koi_yes.loc[index, 'koi_cat_score'] = 30  # Score for "Cold Planet"
    else:
        koi_yes.loc[index, 'koi_cat_score'] = 0  # Score for "Unknown" if any condition is unmet

# Display the DataFrame with scores
print(koi_yes)
```

```python
count=koi_yes['koi_cat_score'].value_counts()
print(count)
plt.figure(figsize=(10,10))
plt.pie(count,labels=count.index,autopct='%1.1f%%')
plt.show()
```

```python
# Initialize an empty list for scores based on koi_srad values\
#Total score 90
srad_scores = []

# Iterate through each value in koi_srad and assign scores
for i in koi_yes['koi_srad']:
    if i == 1:
        srad_scores.append(90)  # Score for "In Habitable Zone"
    elif i > 1:
        srad_scores.append(50)  # Score for "Not in Habitable Zone, Hot Planet"
    else:
        srad_scores.append(50)  # Score for "Not in Habitable Zone, Cold Planet"

# Add the scores to the DataFrame in a new column
koi_yes['koi_cat_srad_score'] = srad_scores

# Display the DataFrame to verify
print(koi_yes)
```

```python
# Initialize an empty list for scores based on koi_prad values
#Total scores 70
prad_scores = []

# Iterate through each value in koi_prad and assign scores
for i in koi_yes['koi_prad']:
    if i < 0.5:
        prad_scores.append(20)  # Score for "Smaller than Earth, Rocky (Unlikely to be habitable)"
    elif 0.5 <= i <= 1.5:
        prad_scores.append(70)  # Score for "Earth-sized, Earth-like (Likely to be habitable)"
    elif 1.5 < i <= 2.5:
        prad_scores.append(50)  # Score for "A bit larger than Earth, Super Earth (Likely to be habitable)"
    elif 2.5 < i <= 4:
        prad_scores.append(30)  # Score for "Larger than Earth, likely Gas Giant (Unlikely to be habitable)"
    elif i > 4:
        prad_scores.append(20)  # Score for "Massive like Jupiter, Gas Giants (Unlikely to be habitable)"
    else:
        prad_scores.append(0)   # Score for "Unknown" if any condition is unmet

# Add the scores to the DataFrame in a new column
koi_yes['koi_cat_prad_score'] = prad_scores

# Display the DataFrame to verify
print(koi_yes)
```

```python
plt.hist(koi_yes['koi_insol'])
plt.axvline(koi_yes['koi_insol'].mean(),color='red')
plt.show()
print(koi_yes['koi_insol'].mean())
```

```python
# Initialize an empty list for scores based on koi_insol values
#total score 70
insol_scores = []

# Iterate through each value in koi_insol and assign scores
for i in koi_yes['koi_insol']:
    if i < 0.25:
        insol_scores.append(40)  # Score for "Mars-like, Cold Planet (Unlikely to be habitable)"
    elif 0.25 <= i < 0.8:
        insol_scores.append(60)  # Score for "Earth-like, Warm (Less Likely to be habitable)"
    elif 0.8 <= i <= 1:
        insol_scores.append(70)  # Score for "Earth-like, Warm (Very Likely to be habitable)"
    elif 1 < i < 1.3:
        insol_scores.append(30)  # Score for "Hotter than Earth, Hot Planet (Less Likely to be habitable)"
    elif i >= 1.3:
        insol_scores.append(10)  # Score for "Too Hotter than Earth, Very Hot Planet (Very Less Likely to be habitable)"
    else:
        insol_scores.append(0)   # Score for "Unknown" if any condition is unmet

# Add the scores to the DataFrame in a new column
koi_yes['koi_cat_insol_score'] = insol_scores

# Display the DataFrame to verify
print(koi_yes)
```

```python
koi_yes['koi_cat_insol_score'].value_counts().plot(kind='bar')
plt.show()
```

```python
koi_hab_score=[]
for index in koi_yes.index: # Iterate using index to access rows
  score = ((koi_yes.loc[index, 'koi_cat_score'] + koi_yes.loc[index, 'koi_cat_srad_score'] +
           koi_yes.loc[index, 'koi_cat_prad_score'] + koi_yes.loc[index, 'koi_cat_insol_score']) / 300 )*100
  koi_hab_score.append(score) # Append calculated score to the list

koi_yes['habitability_score']=koi_hab_score # Assign list to new column
```

```python
from sklearn.linear_model import LinearRegression # Corrected the import statement to use sklearn.linear_model instead of sklearn.LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix # Corrected the typo in the module name from prepeocessing to preprocessing
```

```python
x=koi_yes[['koi_cat_score','koi_cat_srad_score','koi_cat_prad_score','koi_cat_insol_score']]
y=koi_yes['habitability_score']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
# Now apply StandardScaler to the data
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
model = LinearRegression()
model.fit(x_train_scaled, y_train) # Use scaled data for training
y_pred = model.predict(x_test_scaled) # Use scaled data for prediction
R2=r2_score(y_test,y_pred)
print(R2)
mse = mean_squared_error(y_test, y_pred)
print(mse)
```
