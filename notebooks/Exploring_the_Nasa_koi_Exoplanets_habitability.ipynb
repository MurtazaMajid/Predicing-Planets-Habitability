# Computing scores

Computing steff score - 0 to 100

```python
# Initialize the score column
koi_yes['koi_steff_score'] = 0.0

# Sun-like ideal stellar temperature (Kelvin)
ideal_center = 5778   # Sun's effective temperature
spread = 800          # Controls tolerance range

# Iterate and assign continuous Gaussian-based scores
for index, temp in koi_yes['koi_steff'].items():

    # Gaussian scoring: closer to Sun-like temperature â†’ higher score
    score = 100 * np.exp(-((temp - ideal_center) ** 2) / (2 * spread ** 2))

    # Ensure score stays between 0 and 100
    score = max(0, min(100, score))

    koi_yes.loc[index, 'koi_steff_score'] = round(score, 2)

print(koi_yes[['koi_steff', 'koi_steff_score']].head())
```

Computing srad score - 0 to 100

```python
# Ideal value for koi_srad (Earth = 1 solar radius equivalent)
ideal = 1.0
spread = 0.5  # sensitivity around the ideal value

srad_scores = []

# Iterate through koi_srad values and assign continuous scores
for i in koi_yes['koi_srad']:
    # Gaussian-like scoring: closer to 1 â†’ higher score
    score = 90 * np.exp(-((i - ideal) ** 2) / (2 * spread ** 2))

    score = max(0, min(90, score))

    srad_scores.append(round(score, 2))


koi_yes['koi_srad_score'] = srad_scores


print(koi_yes[['koi_srad', 'koi_srad_score']].head())
```

Computing prad score - 0 to 100

```python
def prad_score_continuous(prad):
    # Ideal radius = 1 (Earth radius)
    ideal = 1.0
    # Width factor (controls tolerance, here ~0.8 RâŠ•)
    sigma = 0.8

    # Gaussian decay from the ideal value
    score = 70 * np.exp(-0.5 * ((prad - ideal) / sigma) ** 2)

    # Ensure score does not go negative
    return max(0, score)


koi_yes['koi_prad_score'] = koi_yes['koi_prad'].apply(prad_score_continuous)

print(koi_yes[['koi_prad', 'koi_prad_score']].head(10))
```

Computing insol score - 0 to 100

```python
def insol_score_continuous(insol):
    """
    Continuous habitability score based on stellar flux.
    Uses Gaussian centered at Earth flux (1.0) in log-space.
    """

    # Remove invalid values
    if insol <= 0:
        return 0

    ideal = 1.0     # Earth-like stellar flux
    sigma = 0.5     # Width of habitable zone (0.4â€“0.6 reasonable)

    # Log-space difference
    log_diff = np.log(insol / ideal)

    # Gaussian score (scaled 0â€“70 like your example)
    score = 70 * np.exp(- (log_diff ** 2) / (2 * sigma ** 2))

    return max(0, score)


# Apply to dataframe
koi_yes = koi_yes[koi_yes['koi_insol'] > 0]  # remove invalid rows safely
koi_yes['koi_insol_score'] = koi_yes['koi_insol'].apply(insol_score_continuous)

print(koi_yes[['koi_insol', 'koi_insol_score']].head(10))
```

Computing teq score - 0 to 70

```python
def teq_continuous(teq, ideal=288, sigma=0.4, max_points=70):

    if teq <= 0 or np.isnan(teq):
        return 0

    log_diff = np.log(teq / ideal)

    score = max_points * np.exp(- (log_diff ** 2) / (2 * sigma ** 2))

    return score
koi_yes['koi_teq_score'] = koi_yes['koi_teq'].apply(lambda x: teq_continuous(x))
print(koi_yes[['koi_teq', 'koi_teq_score']].head())
```

Computing orbital period score

```python
def score_period_continuous(series_period, max_points=50, period_ideal=365.0, period_sigma_log=np.log(2)):
    s = pd.to_numeric(series_period, errors='coerce')
    s = s.where(s > 0, np.nan)

    # Distance in log space (so 180 and 730 days are equally "off" from 365)
    d = np.log(s / period_ideal)
    raw = np.exp(-0.5 * (d / period_sigma_log) ** 2)  # Gaussian decay
    scores = max_points * raw
    return scores.fillna(0.0)

koi_yes['koi_period_score'] = score_period_continuous(koi_yes['koi_period'])
print(koi_yes[['koi_period', 'koi_period_score']].head())
```

# **Model Training**

```python
koi_yes_selected=koi_yes[['koi_steff_score', 'koi_srad_score', 'koi_prad_score',
               'koi_insol_score', 'koi_period_score', 'koi_teq_score',
               'habitability_score']]
```

```python
import pandas as pd
import xgboost as xgb
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import r2_score, mean_squared_error
from scipy.stats import uniform, randint
import joblib


# 1ï¸âƒ£ Features & target
X = koi_yes_selected.drop('habitability_score', axis=1)
y = koi_yes_selected['habitability_score']

# 2ï¸âƒ£ Scale features
feature_scaler = MinMaxScaler()
X_scaled = pd.DataFrame(
    feature_scaler.fit_transform(X),
    columns=X.columns,
    index=X.index
)

# 3ï¸âƒ£ Train-test split
x_train, x_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# 4ï¸âƒ£ Define XGBoost regressor
xgb_model = xgb.XGBRegressor(
    objective='reg:squarederror',
    random_state=42,
    n_jobs=-1
)

# 5ï¸âƒ£ Define hyperparameter search space
param_dist = {
    'n_estimators': randint(400, 1200),
    'max_depth': randint(3, 8),
    'learning_rate': uniform(0.01, 0.1),
    'subsample': uniform(0.7, 0.2),
    'colsample_bytree': uniform(0.6, 0.3),
    'reg_lambda': uniform(0.1, 0.5),
    'reg_alpha': uniform(0.0, 0.3),
    'gamma': uniform(0.0, 0.1)
}

# 6ï¸âƒ£ Randomized Search CV
random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_dist,
    n_iter=50,                # number of parameter combinations to try
    scoring='neg_mean_squared_error',
    cv=5,                     # 5-fold CV
    verbose=2,
    random_state=42,
    n_jobs=-1
)

# 7ï¸âƒ£ Fit randomized search
random_search.fit(x_train, y_train)

# 8ï¸âƒ£ Best model from search
best_model = random_search.best_estimator_

# 9ï¸âƒ£ Predict & evaluate
y_pred = best_model.predict(x_test)
print("Best Parameters:", random_search.best_params_)
print("R2:", r2_score(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
```

```python
# Features & target
X = koi_yes_selected.drop('habitability_score', axis=1)
y = koi_yes_selected['habitability_score']

# Scale features
feature_scaler = MinMaxScaler()
X_scaled = pd.DataFrame(
    feature_scaler.fit_transform(X),
    columns=X.columns,
    index=X.index
)

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# XGBoost with tuned hyperparameters
xgb_model = xgb.XGBRegressor(
    objective='reg:squarederror',
    n_estimators=1087,
    learning_rate=0.075761,
    max_depth=3,
    subsample=0.75304,
    colsample_bytree=0.79489,
    reg_alpha=0.0281,
    reg_lambda=0.28386,
    gamma=0.08492,
    random_state=42,
    n_jobs=-1
)

# Train the model
xgb_model.fit(x_train, y_train)

# Evaluate
y_pred = xgb_model.predict(x_test)
print("R2:", r2_score(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))

# Save model and scaler
joblib.dump(xgb_model, "xgb_habitability_model.pkl")
joblib.dump(feature_scaler, "feature_scaler.pkl")
```

Testing the model on earths values for the important features


```python
# Load trained model and scaler
xgb_model = joblib.load("xgb_habitability_model.pkl")
feature_scaler = joblib.load("feature_scaler.pkl")

# Define new planets (raw scores)
test_planets = pd.DataFrame([
    {'Planet': 'Earth',      'koi_steff_score': 65, 'koi_srad_score': 85, 'koi_prad_score': 65, 'koi_insol_score': 65, 'koi_period_score': 30, 'koi_teq_score': 70},
    {'Planet': 'Kepler-22b', 'koi_steff_score': 60, 'koi_srad_score': 80, 'koi_prad_score': 50, 'koi_insol_score': 60, 'koi_period_score': 40, 'koi_teq_score': 60},
    {'Planet': 'Mars',       'koi_steff_score': 50, 'koi_srad_score': 60, 'koi_prad_score': 60, 'koi_insol_score': 40, 'koi_period_score': 50, 'koi_teq_score': 40}
])

planet_names = test_planets['Planet']
planet_features = test_planets.drop(columns=['Planet'])

# Scale features
planet_scaled = pd.DataFrame(
    feature_scaler.transform(planet_features),
    columns=planet_features.columns
)

# Predict scores
predictions = xgb_model.predict(planet_scaled)

# Label habitability relative to Earth
earth_score = predictions[planet_names.tolist().index('Earth')]

def label_habitability(score, earth_score):
    pct = score / earth_score
    if pct >= 0.95:
        return "Habitable"
    elif pct >= 0.8:
        return "Possibly Habitable"
    else:
        return "Uninhabitable"

labels = [label_habitability(s, earth_score) for s in predictions]

# Combine results
results = pd.DataFrame({
    'Planet': planet_names,
    'Predicted Habitability Score': predictions,
    'Habitability Label': labels
})

print(results)
```
