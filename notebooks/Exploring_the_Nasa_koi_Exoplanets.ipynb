# Computing scores

Computing steff score - 0 to 100

```python
# Initialize the score column
koi_yes['koi_steff_score'] = 0.0

# Ideal habitable range for koi_steff (in Kelvin, just an example)
ideal_min, ideal_max = 200, 300
ideal_center = (ideal_min + ideal_max) / 2   # midpoint ~250
spread = 150  # how wide the "habitable zone" is in scoring

# Iterate and assign continuous scores
for index, i in koi_yes['koi_steff'].items():
    # Gaussian-like scoring: closer to 250 â†’ higher score
    score = 100 * np.exp(-((i - ideal_center) ** 2) / (2 * spread ** 2))

    score = max(0, min(100, score))

    koi_yes.loc[index, 'koi_steff_score'] = round(score, 2)

print(koi_yes[['koi_steff', 'koi_steff_score']].head())
```

Computing srad score - 0 to 100

```python
# Ideal value for koi_srad (Earth = 1 solar radius equivalent)
ideal = 1.0
spread = 0.5  # sensitivity around the ideal value

srad_scores = []

# Iterate through koi_srad values and assign continuous scores
for i in koi_yes['koi_srad']:
    # Gaussian-like scoring: closer to 1 â†’ higher score
    score = 90 * np.exp(-((i - ideal) ** 2) / (2 * spread ** 2))

    score = max(0, min(90, score))

    srad_scores.append(round(score, 2))


koi_yes['koi_srad_score'] = srad_scores


print(koi_yes[['koi_srad', 'koi_srad_score']].head())
```

Computing prad score - 0 to 100

```python
def prad_score_continuous(prad):
    # Ideal radius = 1 (Earth radius)
    ideal = 1.0
    # Width factor (controls tolerance, here ~0.8 RâŠ•)
    sigma = 0.8

    # Gaussian decay from the ideal value
    score = 70 * np.exp(-0.5 * ((prad - ideal) / sigma) ** 2)

    # Ensure score does not go negative
    return max(0, score)


koi_yes['koi_prad_score'] = koi_yes['koi_prad'].apply(prad_score_continuous)

print(koi_yes[['koi_prad', 'koi_prad_score']].head(10))
```

Computing insol score - 0 to 100

```python
def insol_score_continuous(insol):
    ideal = 1.0       # Earth-like stellar flux
    sigma = 0.3       # Width factor: how tolerant we are of deviations

    # Gaussian curve around ideal
    score = 70 * np.exp(-0.5 * ((insol - ideal) / sigma) ** 2)

    return max(0, score)  # Keep scores non-negative

koi_yes['koi_insol_score'] = koi_yes['koi_insol'].apply(insol_score_continuous)

print(koi_yes[['koi_insol', 'koi_insol_score']].head(10))
```

Computing teq score - 0 to 70

```python
def teq_continuous(teq, ideal=288, sigma=40, max_points=70, min_points=20):
    """
    Continuous scoring for equilibrium temperature.
    - ideal: Earth-like equilibrium temperature (K)
    - sigma: tolerance (spread) in K
    - max_points: score at ideal temperature
    - min_points: floor score for very far values
    """
    if np.isnan(teq):
        return 0
    # Gaussian curve centered at ideal temp
    score = max_points * np.exp(-0.5 * ((teq - ideal) / sigma) ** 2)
    # Keep at least min_points
    return max(score, min_points)

koi_yes['koi_teq_score'] = koi_yes['koi_teq'].apply(lambda x: teq_continuous(x))
print(koi_yes[['koi_teq', 'koi_teq_score']].head())
```

Computing orbital period score

```python
def score_period_continuous(series_period, max_points=50, period_ideal=365.0, period_sigma_log=np.log(2)):
    s = pd.to_numeric(series_period, errors='coerce')
    s = s.where(s > 0, np.nan)

    # Distance in log space (so 180 and 730 days are equally "off" from 365)
    d = np.log(s / period_ideal)
    raw = np.exp(-0.5 * (d / period_sigma_log) ** 2)  # Gaussian decay
    scores = max_points * raw
    return scores.fillna(0.0)

koi_yes['koi_period_score'] = score_period_continuous(koi_yes['koi_period'])
print(koi_yes[['koi_period', 'koi_period_score']].head())
```

Computing the Habitability score

```python
max_total = 330 + 50 + 70  # 330 (previous 4) + 50 (period) + 70 (teq) = 450

koi_yes['habitability_score'] = (
    (koi_yes['koi_steff_score'] +
     koi_yes['koi_srad_score'] +
     koi_yes['koi_prad_score'] +
     koi_yes['koi_insol_score'] +
     koi_yes['koi_period_score'] +
     koi_yes['koi_teq_score']) / max_total) * 100

print(koi_yes[['koi_steff_score', 'koi_srad_score', 'koi_prad_score',
               'koi_insol_score', 'koi_period_score', 'koi_teq_score',
               'habitability_score']].head())
```

```python
# habitability categories
def categorize_habitability(score):
    if score >= 70:
        return "High Habitability"
    elif 40 <= score < 70:
        return "Medium Habitability"
    else:
        return "Low Habitability"

koi_yes['habitability_category'] = koi_yes['habitability_score'].apply(categorize_habitability)

plt.figure(figsize=(8,5))
sns.countplot(data=koi_yes, x='habitability_category', order=["Low Habitability", "Medium Habitability", "High Habitability"], palette="viridis")
plt.title("Distribution of Exoplanets by Habitability Score Category", fontsize=14)
plt.xlabel("Habitability Category", fontsize=12)
plt.ylabel("Count of Exoplanets", fontsize=12)
plt.show()
```

# **Model Training**

```python
koi_yes_selected=koi_yes[['koi_steff_score', 'koi_srad_score', 'koi_prad_score',
               'koi_insol_score', 'koi_period_score', 'koi_teq_score',
               'habitability_score']]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
koi_yes_selected_scaled = pd.DataFrame(
    scaler.fit_transform(koi_yes_selected),
    columns=koi_yes_selected.columns,
    index=koi_yes_selected.index
)

koi_yes_selected_scaled.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
```

```python
x=koi_yes_selected_scaled.drop('habitability_score',axis=1)
y=koi_yes_selected_scaled['habitability_score']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
model = LinearRegression()
model.fit(x_train_scaled, y_train)
y_pred = model.predict(x_test_scaled)
R2=r2_score(y_test,y_pred)
print(R2)
mse = mean_squared_error(y_test, y_pred)
print(mse)
```
